```python
#######################################################################
# Copyright (C)                                                       #
# 2019 Renault SW Labs                                                #
# Permission given to modify the code as long as you keep this        #
# declaration at the top                                              #
#######################################################################

import numpy as np
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt
from matplotlib import cm

def create_scheduling_list():
    schedule_list = []
    # entry :
    # entry name, type, duration, number taks triggered, [task number trigger, core, 40 ms dependency], [task number trigger, core, 10/40 ms dependency], ...
    # time is expressed in useconds
    # 0
    OS_it_cost_per_task = 200
    CAN_ETH_diag_cost = 1500
    full_list = []
    entry_list = []
    entry_list.append("OS 10 ms tick")
    entry_list.append(0)
    entry_list.append(0)
    entry_list.append(1)
    task_list = []
    task_list.append(1)
    task_list.append(0)
    task_list.append(0)
    entry_list.append(task_list)
    full_list.append(entry_list)
    # 1
    entry_list = []
    entry_list.append("Sup_IN_Rte")
    entry_list.append(1)
    entry_list.append(np.random.choice([600 + OS_it_cost_per_task, 700 + OS_it_cost_per_task, 1100 + OS_it_cost_per_task]))
    entry_list.append(4)
    task_list = []
    task_list.append(2)
    task_list.append(0)
    task_list.append(0)
    entry_list.append(task_list)
    task_list = []
    task_list.append(7)
    task_list.append(2)
    task_list.append(0)
    entry_list.append(task_list)
    task_list = []
    task_list.append(16)
    task_list.append(1)
    task_list.append(1)
    entry_list.append(task_list)
    task_list = []
    task_list.append(14)
    task_list.append(3)
    task_list.append(0)
    entry_list.append(task_list)
    full_list.append(entry_list)
    # 2
    entry_list = []
    entry_list.append("Sup_OUT")
    entry_list.append(1)
    entry_list.append(np.random.choice([100 + OS_it_cost_per_task, 100 + OS_it_cost_per_task, 200 + OS_it_cost_per_task]))
    entry_list.append(1)
    task_list = []
    task_list.append(3)
    task_list.append(0)
    task_list.append(0)
    entry_list.append(task_list)
    full_list.append(entry_list)
    # 3
    entry_list = []
    entry_list.append("Sup_OUT(BSW)")
    entry_list.append(1)
    entry_list.append(np.random.choice([100 + OS_it_cost_per_task, 100 + OS_it_cost_per_task, 200 + OS_it_cost_per_task]))
    entry_list.append(2)
    task_list = []
    task_list.append(4)
    task_list.append(0)
    task_list.append(0)
    entry_list.append(task_list)
    task_list = []
    task_list.append(13)
    task_list.append(2)
    task_list.append(0)
    entry_list.append(task_list)
    full_list.append(entry_list)
    # 4
    entry_list = []
    entry_list.append("BSW_Out")
    entry_list.append(1)
    entry_list.append(np.random.choice([400 + OS_it_cost_per_task, 500 + OS_it_cost_per_task, 700 + OS_it_cost_per_task]))
    entry_list.append(1)
    task_list = []
    task_list.append(5)
    task_list.append(0)
    task_list.append(0)
    entry_list.append(task_list)
    full_list.append(entry_list)
    # 5
    entry_list = []
    entry_list.append("BSW_IN")
    entry_list.append(1)
    entry_list.append(np.random.choice([700 + OS_it_cost_per_task, 700 + OS_it_cost_per_task, 1000 + OS_it_cost_per_task]))
    entry_list.append(1)
    task_list = []
    task_list.append(6)
    task_list.append(0)
    task_list.append(0)
    entry_list.append(task_list)
    full_list.append(entry_list)
    # 6
    entry_list = []
    entry_list.append("SUP_IN")
    entry_list.append(1)
    entry_list.append(np.random.choice([2700 + OS_it_cost_per_task + CAN_ETH_diag_cost, 2700 + OS_it_cost_per_task + CAN_ETH_diag_cost, 4100 + OS_it_cost_per_task + CAN_ETH_diag_cost]))
    entry_list.append(0)
    task_list = []
    entry_list.append(task_list)
    full_list.append(entry_list)
    # 7
    entry_list = []
    entry_list.append("OEM1_1")
    entry_list.append(1)
    entry_list.append(np.random.choice([100, 100, 200]))
    entry_list.append(1)
    task_list = []
    task_list.append(8)
    task_list.append(2)
    task_list.append(0)
    entry_list.append(task_list)
    full_list.append(entry_list)
    # 8
    entry_list = []
    entry_list.append("OEM1_2")
    entry_list.append(1)
    entry_list.append(np.random.choice([300, 300, 500]))
    entry_list.append(1)
    task_list = []
    task_list.append(9)
    task_list.append(2)
    task_list.append(0)
    entry_list.append(task_list)
    full_list.append(entry_list)
    # 9
    entry_list = []
    entry_list.append("OEM2")
    entry_list.append(1)
    entry_list.append(np.random.choice([1600, 1600, 2400]))
    entry_list.append(3)
    task_list = []
    task_list.append(1)
    task_list.append(0)
    task_list.append(0)
    entry_list.append(task_list)
    task_list = []
    task_list.append(10)
    task_list.append(2)
    task_list.append(0)
    entry_list.append(task_list)
    task_list = []
    task_list.append(15)
    task_list.append(3)
    task_list.append(0)
    entry_list.append(task_list)
    full_list.append(entry_list)
    # 10
    entry_list = []
    entry_list.append("OEM3_1")
    entry_list.append(1)
    entry_list.append(np.random.choice([400, 400, 600]))
    entry_list.append(1)
    task_list = []
    task_list.append(11)
    task_list.append(2)
    task_list.append(0)
    entry_list.append(task_list)
    full_list.append(entry_list)
    # 11
    entry_list = []
    entry_list.append("OEM3_2")
    entry_list.append(1)
    entry_list.append(np.random.choice([500, 500, 800]))
    entry_list.append(1)
    task_list = []
    task_list.append(12)
    task_list.append(2)
    task_list.append(0)
    entry_list.append(task_list)
    full_list.append(entry_list)
    # 12
    entry_list = []
    entry_list.append("OEM3_3")
    entry_list.append(1)
    entry_list.append(np.random.choice([500, 500, 800]))
    entry_list.append(1)
    task_list = []
    task_list.append(13)
    task_list.append(2)
    task_list.append(0)
    entry_list.append(task_list)
    full_list.append(entry_list)
    # 13
    entry_list = []
    entry_list.append("OEM4")
    entry_list.append(1)
    entry_list.append(np.random.choice([200, 200, 300]))
    entry_list.append(1)
    task_list = []
    task_list.append(1)
    task_list.append(0)
    task_list.append(0)
    entry_list.append(task_list)
    full_list.append(entry_list)
    # 14
    entry_list = []
    entry_list.append("OEM_DPL_SALC")
    entry_list.append(1)
    entry_list.append(np.random.choice([3000, 3300, 5000]))
    entry_list.append(2)
    task_list = []
    task_list.append(12)
    task_list.append(2)
    task_list.append(0)
    entry_list.append(task_list)
    task_list = []
    task_list.append(15)
    task_list.append(3)
    task_list.append(0)
    entry_list.append(task_list)
    full_list.append(entry_list)
    # 15
    entry_list = []
    entry_list.append("OEM_AEB")
    entry_list.append(1)
    entry_list.append(np.random.choice([1500, 1500, 2300]))
    entry_list.append(1)
    task_list = []
    task_list.append(13)
    task_list.append(2)
    task_list.append(0)
    entry_list.append(task_list)
    full_list.append(entry_list)
    # 16
    entry_list = []
    entry_list.append("FusionIn")
    entry_list.append(1)
    entry_list.append(np.random.choice([100, 100, 200]))
    entry_list.append(1)
    task_list = []
    task_list.append(17)
    task_list.append(1)
    task_list.append(1)
    entry_list.append(task_list)
    full_list.append(entry_list)
    # 17
    entry_list = []
    entry_list.append("FusionRT")
    entry_list.append(1)
    entry_list.append(np.random.choice([10000, 22000, 30000]))
    entry_list.append(1)
    task_list = []
    task_list.append(18)
    task_list.append(1)
    task_list.append(1)
    entry_list.append(task_list)
    full_list.append(entry_list)
    # 18
    entry_list = []
    entry_list.append("FusionOut")
    entry_list.append(1)
    entry_list.append(np.random.choice([100, 100, 200]))
    entry_list.append(1)
    task_list = []
    task_list.append(1)
    task_list.append(0)
    task_list.append(0)
    entry_list.append(task_list)
    full_list.append(entry_list)

    return full_list

def determine_dependency_flags(task_num, full_list, list_length):
    task_dependency_list = []
    flag_list = []
    occ = 0
    for i in range (list_length - 1):
        idx = i + 1
        for j in range(full_list[idx][3]):
            if full_list[idx][4 + j][0] == task_num:
                task_dependency_list.append(idx)
                occ = occ + 1
    flag_list. append(occ)
    flag_list.append(task_dependency_list)
    return flag_list

def get_task(task_num, full_list, list_length):
    dispatch_list = []
    if full_list[task_num][1] == 1:
        nb_dispatch = full_list[task_num][3]
        nb_dispatch1 = nb_dispatch
        for i in range(nb_dispatch):
            task_to_launch = full_list[task_num][4 + i][0]
            if task_to_launch == 1:
                nb_dispatch1 = nb_dispatch1 - 1
        dispatch_list.append(nb_dispatch1)
        for i in range(nb_dispatch):
            local_list = []
            task_to_launch = full_list[task_num][4 + i][0]
            if task_to_launch != 1:
                local_core = full_list[task_num][4 + i][1]
                # task name
                local_list.append(full_list[task_to_launch][0])
                # task number
                local_list.append(task_to_launch)
                # core number
                local_list.append(local_core)
                # duration
                local_list.append(full_list[task_to_launch][2])
                # dependencies
                local_list.append(determine_dependency_flags(task_to_launch, full_list, list_length))
                dispatch_list.append(local_list)
    else:
        nb_dispatch = 1
        dispatch_list.append(nb_dispatch)
        local_list = []
        task_to_launch = full_list[task_num][4][0]
        local_core = full_list[task_num][4][1]
        # task name
        local_list.append(full_list[task_to_launch][0])
        # task number
        local_list.append(task_to_launch)
        # core number
        local_list.append(local_core)
        # duration
        local_list.append(full_list[task_to_launch][2])
        # dependencies
        local_list.append(determine_dependency_flags(task_to_launch, full_list, list_length))
        dispatch_list.append(local_list)

    return dispatch_list

def set_all_flags(list_length):
    flag_list = []
    for i in range(list_length):
        flag_list.append(1)

    return flag_list

def clear_all_flags(list_length):
    flag_list = []
    for i in range(list_length):
        flag_list.append(0)

    return flag_list

def clear_all_times(list_length):
    flag_list = []
    for i in range(list_length):
        flag_list.append(0)

    return flag_list

def launch_tasks(launch_list, core, tasks_flags, tasks_times, tasks_ticks, time_at_core_0, t, ef):
    for i in range(launch_list[0]):
        flg_count = 0
        for j in range(launch_list[1 + i][4][0]):
            flg_idx = launch_list[1 + i][4][1][j]
            if tasks_ticks[flg_idx] == 40:
                if time_at_core_0 >= t:
                    valid_flag = 1
                else:
                    valid_flag = 0
            else:
                valid_flag = 1
            if tasks_flags[flg_idx] == 1 and tasks_times[flg_idx] <= t and valid_flag == 1 or ef == 1:
                flg_count = flg_count + 1
        if flg_count == launch_list[1 + i][4][0] and core == launch_list[1 + i][2]:
            tasks_flags[launch_list[1 + i][1]] = 0
            print("Launching task ", launch_list[1 + i][0], " on core ", launch_list[1 + i][2])

    return tasks_flags

def determine_waiting_tasks(core, launch_list, tasks_flags, tasks_times, tasks_ticks, t_10ms_flag, t_40ms_flag, t):
    wait_list = []
    full_wait_list = []
    wait_count = 0
    time_2_wait = 0
    for i in range(launch_list[0]):
        flg_count = 0
        for j in range(launch_list[1 + i][4][0]):
            flg_idx = launch_list[1 + i][4][1][j]
            if (tasks_ticks[flg_idx] == 40 and (t_40ms_flag[1] == 1 or t < tasks_times[flg_idx])) or (tasks_ticks[flg_idx] == 10 and t_10ms_flag[core] == 1):
                valid_test = 1
            else:
                valid_test = 0
            if tasks_flags[flg_idx] == 1 and (tasks_times[flg_idx] <= t or valid_test == 1 or (t_40ms_flag[1] == 1 and (t_10ms_flag[core] == 1 or core == 1))):
                flg_count = flg_count + 1
            else:
                if core == launch_list[1 + i][2]:
                    print("Task", flg_idx, "not ready!")
                    if tasks_times[flg_idx] > t:
                        time_2_wait = tasks_times[flg_idx] - t
                        print("Time to wait for this task =", time_2_wait)
        if flg_count != launch_list[1 + i][4][0] and core == launch_list[1 + i][2]:
            wait_count = wait_count + 1
            wait_list.append(launch_list[1 + i][1])
            wait_list.append(launch_list[1 + i][2])
            wait_list.append(time_2_wait)
    full_wait_list.append(wait_count)
    full_wait_list.append(wait_list)

    return full_wait_list

def determine_next_time_events(core, time, launch_list, tasks_flags, tasks_times, tasks_ticks, t_10ms_flag, t_40ms_flag):
    time_list = []
    full_time_list = []
    time_count = 0
    for i in range(launch_list[0]):
        flg_count = 0
        for j in range(launch_list[1 + i][4][0]):
            flg_idx = launch_list[1 + i][4][1][j]
            if (tasks_ticks[flg_idx] == 40 and (t_40ms_flag[1] == 1 or time < tasks_times[flg_idx])) or(tasks_ticks[flg_idx] == 10 and t_10ms_flag[core] == 1):
                valid_test = 1
            else:
                valid_test = 0
            if tasks_flags[flg_idx] == 1 and (tasks_times[flg_idx] <= time or valid_test == 1 or t_40ms_flag[1] == 1):
                flg_count = flg_count + 1
        if flg_count == launch_list[1 + i][4][0] and core == launch_list[1 + i][2]:
            t = time + launch_list[1 + i][3]
            time_count = time_count + 1
            time_list.append(t)
            time_list.append(launch_list[1 + i][2])
            time_list.append(launch_list[1 + i][1])
    full_time_list.append(time_count)
    full_time_list.append(time_list)

    return full_time_list

def go_next(next_time_events):
    out_list = []
    for i in range(next_time_events[0]):
        time = next_time_events[1 + i][0]
        if i == 0:
            min_time = time
            c = next_time_events[1 + i][1]
            task = next_time_events[1 + i][2]
            idx = 0
        else:
            if time < min_time:
                min_time = time
                c = next_time_events[1 + i][1]
                task = next_time_events[1 + i][2]
                idx = 0
    out_list.append(min_time)
    out_list.append(c)
    out_list.append(idx + 1)
    out_list.append(task)

    return out_list

def clean_events(next_time_events, idx):
    del next_time_events[idx]
    next_time_events[0] = next_time_events[0] - 1

    return next_time_events

def close_task(task, tasks_flags):
    tasks_flags[task] = 1

    return tasks_flags

def clear_all_core_flags(tasks_list, tasks_flags):
    for i in range (len(tasks_list)):
        tasks_flags[tasks_list[i]] = 0

    return tasks_flags

def task_to_launch(launch_list, task_idx):
    return [launch_list[1 + task_idx][1], launch_list[1 + task_idx][2]]

def core_run(task, tasks_flags, tasks_times, tasks_ticks, t_10ms_flag, t_40ms_flag, t, core, cores_active, T_list, Cores_tasks_list, Launch_dependency, count_40ms_slots, time_map):
    core_results = []
    launch_list = get_task(task, T_list, 19)
    #print(launch_list)
    if ((t[0] - count_40ms_slots[1] * 40000) <= 10000):
        next_40ms_time = 0
    else:
        next_40ms_time = 40000 + (t[0] - t[0] % 40000)
    waiting_tasks = determine_waiting_tasks(core, launch_list, tasks_flags, tasks_times, tasks_ticks, t_10ms_flag, t_40ms_flag, t[core])
    #print(waiting_tasks)
    # Fixing the issue of 40 ms task time already being at near 40 ms while other tasks are at 10 ms slots
    if waiting_tasks[0] == 0:
        events = determine_next_time_events(core, t[core], launch_list, tasks_flags, tasks_times, tasks_ticks, t_10ms_flag, t_40ms_flag)
        #print("Checks for 40 ms: events =", events)
        if events[0] != 0:
            task_id = events[1][2]
            #print("Checks for 40 ms: task_id =", task_id)
            #print("Checks for 40 ms: tasks_ticks =", tasks_ticks[task_id])
            if tasks_ticks[task_id] == 40:
                task_id_40ms = task_id
                if t[0] < max(tasks_times[Cores_tasks_list[core][-1]], next_40ms_time):
                    print("task ID that is to be waiting and not restarting =", task_id_40ms)
                    cores_active[core] = 1
                    filter_out = 1
                else:
                    filter_out = 0
            else:
                filter_out = 0
        else:
            filter_out = 0
    else:
        filter_out = 0
    if waiting_tasks[0] != 0 or filter_out == 1:
        task1 = task
        flg = 0
        if filter_out == 0:
            time_2_wait = waiting_tasks[1][2]
        else:
            time_2_wait = 0
        if filter_out == 1:
            goto[core] = 1
    else:
        next_time_events = determine_next_time_events(core, t[core], launch_list, tasks_flags, tasks_times, tasks_ticks, t_10ms_flag, t_40ms_flag)
        #print(next_time_events)
        if (next_time_events[0] == 0):
            if core != 0:
                last_task_to_exec_on_core = Cores_tasks_list[core][-1]
                if task == last_task_to_exec_on_core:
                    print("Reached end of tasks to exec on this core!")
                    loop_over_flag = 1
                else:
                    print("Have completed my tasks! Last task to cover on this core is", last_task_to_exec_on_core)
                    loop_over_flag = 0
            else:
                loop_over_flag = 1
            if loop_over_flag == 1:
                task1 = Launch_dependency[core]
                flg = 1
                time_2_wait = 0
            else:
                task1 = task
                flg = 0
                time_2_wait = 0
        else:
            cores_active[core] = 1
            #print(cores_active)
            print("At time =", t[core])
            tasks_flags = launch_tasks(launch_list, core, tasks_flags, tasks_times, tasks_ticks, t[0], t[core], 0)
            #print(tasks_flags)
            [t1, c, idx, task] = go_next(next_time_events)
            print("New time =", t1, ", core to free up =", c, ", task to terminate =", task, ", Event to cancel =", idx)
            time_map[core, int(t[core] / 100)] = 10 * task + Tasks_offset_color[task]
            #print("Gfx data:", core, int(t[core] / 100), time_map[core, int(t[core] / 100)])
            time_map[core, int(t1 / 100) - 1] = 10 * task + Tasks_offset_color[task]
            index = int(t[core] / 100) + 1
            while index < (int(t1 / 100) - 1):
                time_map[core, index] = 10 * task + Tasks_offset_color[task]
                #print("Gfx data:", core, index, time_map[core, index])
                index = index + 1
            #if int(t1 / 100) - 1 > int(t[core] / 100):
            #    print("Gfx data:", core, int(t1 / 100) - 1, time_map[core, int(t1 / 100) - 1])
            t[c] = t1
            next_time_events = clean_events(next_time_events, idx)
            #print(next_time_events)
            #print(t)
            #print(cores_active)
            tasks_flags = close_task(task, tasks_flags)
            tasks_times[task] = t1
            if Cores_tasks_list[core][-1] == task:
                task1 = Launch_dependency[core]
            else:
                task1 = task
            flg = 0
            time_2_wait = 0
    core_results.append(task1)
    core_results.append(flg)
    core_results.append(time_2_wait)

    return(core_results)

def set_all_tasks_times_to_next_tick(tasks_times, tasks_ticks, size, tick_val):
    for i in range (size):
        if i != 0:
            if tasks_ticks[i] == tick_val:
                tasks_times[i] = tasks_times[i] + ((tick_val * 1000) - tasks_times[i] % (tick_val * 1000))

    return tasks_times

def set_tasks_flags_for_40ms_tasks(tasks_ticks, tasks_flags, size):
    for i in range(size):
        if tasks_ticks[i] == 40:
            tasks_flags[i] = 1

def define_cores_tasks_list(Tasks_list, size):
    Cores_tasks_list = []
    core0_list = []
    core1_list = []
    core2_list = []
    core3_list = []
    for i in range(size - 1):
        launch_list = get_task(i, Tasks_list, size)
        for j in range(launch_list[0]):
            if launch_list[1 + j][2] == 0:
                if launch_list[1 + j][1] not in core0_list:
                    core0_list.append(launch_list[1 + j][1])
                    core0_list.sort()
            elif launch_list[1 + j][2] == 1:
                if launch_list[1 + j][1] not in core1_list:
                    core1_list.append(launch_list[1 + j][1])
                    core1_list.sort()
            elif launch_list[1 + j][2] == 2:
                if launch_list[1 + j][1] not in core2_list:
                    core2_list.append(launch_list[1 + j][1])
                    core2_list.sort()
            else:
                if launch_list[1 + j][1] not in core3_list:
                    core3_list.append(launch_list[1 + j][1])
                    core3_list.sort()
    Cores_tasks_list.append(core0_list)
    Cores_tasks_list.append(core1_list)
    Cores_tasks_list.append(core2_list)
    Cores_tasks_list.append(core3_list)

    return Cores_tasks_list

def define_tasks_ticks(Tasks_list, size):
    tasks_ticks = []
    for i in range(size):
        tasks_ticks.append(0)
    tasks_ticks[0] = 10
    for i in range(size):
        nb_tasks = Tasks_list[i][3]
        for j in range(nb_tasks):
            task_type = Tasks_list[i][4 + j][2]
            task = Tasks_list[i][4 + j][0]
            if task_type == 0:
                tick = 10
            else:
                tick = 40
            if tasks_ticks[task] == 0:
                tasks_ticks[task] = tick

    return tasks_ticks

def define_GFx_Offsets_list(Tasks_list, size):
    Tasks_offset_color = []
    offset_list = []
    offset_list.append(10)
    offset_list.append(70)
    offset_list.append(30)
    offset_list.append(50)
    offset_list.append(100)
    offset_list.append(120)
    for i in range(size):
        Tasks_offset_color.append(0)
    for i in range(size - 1):
        launch_list = get_task(i, Tasks_list, size)
        for j in range(launch_list[0]):
            if Tasks_offset_color[launch_list[1 + j][1]] == 0:
                Tasks_offset_color[launch_list[1 + j][1]] = int(offset_list[launch_list[1 + j][2]])

    return Tasks_offset_color

def define_GFx_TasksNames_list(Tasks_list, size):
    Tasks_names = []
    for i in range(size - 1):
        Tasks_names.append(0)
    for i in range(size - 1):
        launch_list = get_task(i, Tasks_list, size)
        for j in range(launch_list[0]):
            if Tasks_names[launch_list[1 + j][1] - 1] == 0:
                Tasks_names[launch_list[1 + j][1] - 1] = launch_list[1 + j][0]

    return Tasks_names

def define_GFx_legend_list(Tasks_offset_color, size):
    Legend_vals = np.zeros((1, size - 1))
    for i in range(size):
        if i != 0:
            Legend_vals[0, i - 1] = 10 * i + Tasks_offset_color[i]

    return Legend_vals

def define_launch_depency_list(Tasks_list, Cores_tasks_list, size):
    Launch_dependency = []
    n = len(Cores_tasks_list)
    starting_task_per_core = []
    for i in range(n):
        if i == 0:
            starting_task_per_core.append(Cores_tasks_list[i][1])
        else:
            starting_task_per_core.append(Cores_tasks_list[i][0])
        Launch_dependency.append(0)
    for i in range(size - 1):
        launch_list = get_task(i, Tasks_list, size)
        for j in range(launch_list[0]):
            if launch_list[1 + j][1] in starting_task_per_core:
                if launch_list[1 + j][4][0] == 1:
                    Launch_dependency[launch_list[1 + j][2]] = launch_list[1 + j][4][1][0]
                else:
                    local_dep_list = []
                    for k in range(launch_list[1 + j][4][0]):
                        local_dep_list.append(launch_list[1 + j][4][1][k])
                    Launch_dependency[launch_list[1 + j][2]] = local_dep_list

    return Launch_dependency

def launch_dep_task_is_in_core(T_list, Launch_dependency, tasks_times, core):
    r_list = []
    task_to_depend_on = Launch_dependency[core]
    if T_list[task_to_depend_on][1] == core:
        r_flag = 1
        time_start_ref = 0
    else:
        r_flag = 0
        time_start_ref = tasks_times[task_to_depend_on]
    r_list.append(r_flag)
    r_list.append(time_start_ref)

    return r_list

def measure_cores_load(cores_loads, T_list, Launch_dependency, Cores_tasks_list, tasks_times, tasks_ticks, count_10ms_slots, time_start_ref_40, core_num):
    for i in range(core_num):
        idx_last_task = Cores_tasks_list[i][-1]
        if (tasks_ticks[idx_last_task] == 40 and count_10ms_slots[0] % 4 == 0):
            exe_duration_in_core = (tasks_times[idx_last_task] - time_start_ref_40)
            print("40 ms slot: ref time is", time_start_ref_40, "and corresponding 40 ms processing execution time is", exe_duration_in_core)
            exe_percent_per_core = int(exe_duration_in_core / (tasks_ticks[idx_last_task] * 10))
            cores_loads[i][0] = cores_loads[i][0] + 1
            cores_loads[i][2] = int(((cores_loads[i][0] - 1) * cores_loads[i][2] + exe_percent_per_core) / cores_loads[i][0])
            if exe_percent_per_core < cores_loads[i][1]:
                cores_loads[i][1] = exe_percent_per_core
            if exe_percent_per_core > cores_loads[i][3]:
                cores_loads[i][3] = exe_percent_per_core
        if tasks_ticks[idx_last_task] == 10:
            time_start_ref_list = launch_dep_task_is_in_core(T_list, Launch_dependency, tasks_times, i)
            exe_duration_in_core = (tasks_times[idx_last_task] - time_start_ref_list[1])
            exe_percent_per_core = int(exe_duration_in_core / (tasks_ticks[idx_last_task] * 10))
            cores_loads[i][0] = cores_loads[i][0] + 1
            cores_loads[i][2] = int(((cores_loads[i][0] - 1) * cores_loads[i][2] + exe_percent_per_core) / cores_loads[i][0])
            if exe_percent_per_core < cores_loads[i][1]:
                cores_loads[i][1] = exe_percent_per_core
            if exe_percent_per_core > cores_loads[i][3]:
                cores_loads[i][3] = exe_percent_per_core

    return cores_loads

def measure_in_buffer_valid_lat(input_buffer_release_task, input_buffer_consume_task, buffer_release_time, input_binding_ctr, in_buf_cons_lat, tasks_times):
    out_list = []

    t1 = buffer_release_time
    t2 = t1
    if input_binding_ctr > 0:
        t2 = tasks_times[input_buffer_consume_task]
        delta_t = t2 - t1
        print("Current Input Buffer validity need, in time, is", delta_t, "us")
        in_buf_cons_lat[0] = in_buf_cons_lat[0] + 1
        in_buf_cons_lat[3] = int(((in_buf_cons_lat[0] - 1) * in_buf_cons_lat[3] + delta_t) / in_buf_cons_lat[0])
        if delta_t < in_buf_cons_lat[1]:
            in_buf_cons_lat[1] = delta_t
            in_buf_cons_lat[2] = input_binding_ctr + 1
        if delta_t > in_buf_cons_lat[4]:
            in_buf_cons_lat[4] = delta_t
            in_buf_cons_lat[5] = input_binding_ctr + 1
    t1 = tasks_times[input_buffer_release_task]
    input_binding_ctr = input_binding_ctr + 1
    out_list.append(input_binding_ctr)
    out_list.append(t1)
    out_list.append(t2)
    out_list.append(in_buf_cons_lat)

    return out_list
```

```python
while f_break != 1:
    # Time management
    # List exchange (3 - ranmdom pick)
    list_idx = np.random.choice([0, 1, 2])
    if list_idx == 0:
        T_list = T_list0
    elif list_idx == 1:
        T_list = T_list1
    else:
        T_list = T_list2
    if (t[0] % 10000 == 0 or (t[0] - (count_10ms_slots[0] + 1) * 10000 > 0)) and t_10ms_flag[0] == 0:
        print("Setting 10 ms flag of core", 0)
        t_10ms_flag[0] = 1
        count_10ms_slots[0] = count_10ms_slots[0] + 1
        count_flag_10[0] = 1
    elif t[0] % 10000 >= 100:
        t_10ms_flag[0] = 0
        count_flag_10[0] = 0
    if (t[1] % 40000 == 0 or (t[1] - (count_40ms_slots[1] + 1) * 40000 > 0)) and count_flag_40[1] == 0:
        t_40ms_flag[1] = 1
        count_40ms_slots[1] = count_40ms_slots[1] + 1
        count_flag_40[1] = 1
    elif t[1] % 40000 >= 9900:
        t_40ms_flag[1] = 0
        count_flag_40[1] = 0
    if (t[2] % 10000 == 0 or (t[2] - (count_10ms_slots[2] + 1) * 10000 > 0)) and t_10ms_flag[2] == 0:
        print("Setting 10 ms flag of core", 2)
        t_10ms_flag[2] = 1
        count_10ms_slots[2] = count_10ms_slots[2] + 1
        count_flag_10[2] = 1
    elif t[2] % 10000 >= 100:
        t_10ms_flag[2] = 0
        count_flag_10[2] = 0
    if (t[3] % 10000 == 0 or (t[3] - (count_10ms_slots[3] + 1) * 10000 > 0)) and t_10ms_flag[3] == 0:
        print("Setting 10 ms flag of core", 3)
        t_10ms_flag[3] = 1
        count_10ms_slots[3] = count_10ms_slots[3] + 1
        count_flag_10[3] = 1
    elif t[3] % 10000 >= 100:
        t_10ms_flag[3] = 0
        count_flag_10[3] = 0
    # ####### End of Time management
    if count_core2_end == end_iter:
        f_break = 1
    if cpt_watchd > watchdog_duration:
        print("exiting due to Watchdog!")
        f_break = 1
    cpt_watchd= cpt_watchd + 1
    # ##############################
    #print("10/40 ms Flags status")
    #print(t_10ms_flag, t_40ms_flag)
    #print("10/40 ms slots count")
    #print(count_10ms_slots, count_40ms_slots)

    print("############### Cores iterations #################")
    if tasks_flags[Cores_tasks_list[0][-1]] != 1:
        print("Core 0")
        print(task0)
        [task0, flg0, time_2_wait0] = core_run(task0, tasks_flags, tasks_times, tasks_ticks, t_10ms_flag, t_40ms_flag, t, 0, cores_active, T_list, Cores_tasks_list, Launch_dependency, count_40ms_slots, time_map)
        print(cores_active)
        print(tasks_flags)
        print(tasks_times)
        print(t)
        print(task0)
    else:
        print("Core 0")
        t_rem = t[0] % 10000
        t_rem1 = t[0] - count_10ms_slots[0] * 10000
        #print("t_rem =", t_rem, "t_rem1 =", t_rem1)
        if t_rem == t_rem1 and goto[0] == 0:
            count_core0_end = count_core0_end + 1
            t_corr = 10000 - t_rem
            t[0] = t[0] + t_corr
            print("Going to next 10 ms boundary")
            print(t)
            goto[0] = 1
        else:
            print("ready to loop over")
            goto[0] = 1
    if tasks_flags[Cores_tasks_list[1][-1]] != 1:
        print("Core 1")
        print(task1)
        [task1, flg1, time_2_wait1] = core_run(task1, tasks_flags, tasks_times, tasks_ticks, t_10ms_flag, t_40ms_flag, t, 1, cores_active, T_list, Cores_tasks_list, Launch_dependency, count_40ms_slots, time_map)
        print(cores_active)
        print(tasks_flags)
        print(tasks_times)
        print(t)
        print(task1)
    else:
        print("Core 1")
        t_rem = t[1] % 40000
        t_rem1 = t[1] - count_40ms_slots[1] * 40000
        #print("t_rem =", t_rem, "t_rem1 =", t_rem1)
        if t_rem == t_rem1 and goto[1] == 0 and tasks_ticks[Cores_tasks_list[1][-1]] == 40 and t[0] % 40000 == 0:
            if t[1] - (count_40ms_slots[1] + 1) * 40000 > 0:
                print("ready to loop over")
            else:
                t_corr = 40000 - t_rem
                t[1] = t[1] + t_corr
                print("Going to next 40 ms boundary")
                print(t)
            goto[1] = 1
        else:
            print("ready to loop over")
            goto[1] = 1
    if tasks_flags[Cores_tasks_list[2][-1]] != 1:
        print("Core 2")
        print(task2)
        [task2, flg2, time_2_wait2] = core_run(task2, tasks_flags, tasks_times, tasks_ticks, t_10ms_flag, t_40ms_flag, t, 2, cores_active, T_list, Cores_tasks_list, Launch_dependency, count_40ms_slots, time_map)
        print(cores_active)
        print(tasks_flags)
        print(tasks_times)
        print(t)
        print(task2)
    else:
        print("Core 2")
        t_rem = t[2] % 10000
        t_rem1 = t[2] - count_10ms_slots[2] * 10000
        #print("t_rem =", t_rem, "t_rem1 =", t_rem1)
        if t_rem == t_rem1 and goto[2] == 0:
            count_core2_end = count_core2_end + 1
            #print("Core 2 end processing count =", count_core2_end)
            t_corr = 10000 - t_rem
            t[2] = t[2] + t_corr
            print("Going to next 10 ms boundary")
            print(t)
            goto[2] = 1
        else:
            print("ready to loop over")
            goto[2] = 1
    if tasks_flags[Cores_tasks_list[3][-1]] != 1:
        print("Core 3")
        [task3, flg3, time_2_wait3] = core_run(task3, tasks_flags, tasks_times, tasks_ticks, t_10ms_flag, t_40ms_flag, t, 3, cores_active, T_list, Cores_tasks_list, Launch_dependency, count_40ms_slots, time_map)
        print(cores_active)
        print(tasks_flags)
        print(tasks_times)
        print(t)
        print(task3)
        print(time_2_wait3)
    else:
        print("Core 3")
        t_rem = t[3] % 10000
        t_rem1 = t[3] - count_10ms_slots[3] * 10000
        #print("t_rem =", t_rem, "t_rem1 =", t_rem1)
        if t_rem == t_rem1 and goto[3] == 0:
            count_core3_end = count_core3_end + 1
            t_corr = 10000 - t_rem
            t[3] = t[3] + t_corr
            print("Going to next 10 ms boundary")
            print(t)
            goto[3] = 1
        else:
            print("ready to loop over")
            goto[3] = 1
    min_t = 1000000
    for i in range(4):
        if cores_active[i] == 1 and t[i] < min_t and goto[i] == 0:
            min_t = t[i]
    #print("Min t to align =", min_t)
    if min_t == 1000000:
        min_flg_issue = 1
    else:
        min_flg_issue = 0
    for i in range(4):
        if cores_active[i] == 0 and goto[i] == 0 and min_flg_issue == 0:
            if i == 0:
                local_task_idx = task0
                time_2_add = time_2_wait0
            elif i == 1:
                local_task_idx = task1
                time_2_add = time_2_wait1
            elif i == 2:
                local_task_idx = task2
                time_2_add = time_2_wait2
            else:
                local_task_idx = task3
                time_2_add = time_2_wait3
            if time_2_add == 0:
                if t[i] < min_t:
                    t[i] = min_t
            else:
                print("Core waiting =", i, "task =", local_task_idx + 1)
                t[i] = t[i] + time_2_add
                print("Core", i, "waited until time", t[i])
        elif cores_active[i] == 0 and goto[i] == 0 and min_flg_issue == 1:
            # Last core waiting - Here, core 2, waiting for task Core_AEB or core 3
            if i == 0:
                local_task_idx = task0
                time_2_add = time_2_wait0
            elif i == 1:
                local_task_idx = task1
                time_2_add = time_2_wait1
            elif i == 2:
                local_task_idx = task2
                time_2_add = time_2_wait2
            else:
                local_task_idx = task3
                time_2_add = time_2_wait3
            print("Core waiting =", i, "task =", local_task_idx + 1)
            t[i] = t[i] + time_2_add
            print("Core", i, "waited until time", t[i])
        else:
            cores_active[i] = 0
    print("Final times on cores for this round")
    print(t)
    #print("End of task list status")
    #print(goto)
    if t_10ms_flag == [1, 0, 1, 1]:
        print("############### End of 10 ms slot #################")
        print("Re-start")
        mod_flag = 1
        time_adjust = 0
        for j in range(4):
            if count_10ms_slots[j] % 4 != 0:
                mod_flag = 0
        if mod_flag == 1 and count_flag_40[1] == 0:
            t_40ms_flag[1] = 1
            count_40ms_slots[1] = count_40ms_slots[1] + 1
            count_flag_40[1] = 1
        print("10/40 ms Flags status")
        print(t_10ms_flag, t_40ms_flag)
        print("10/40 ms slots count")
        print(count_10ms_slots, count_40ms_slots)
        tasks_flags[0] = 1
        set_tasks_flags_for_40ms_tasks(tasks_ticks, tasks_flags, 19)
        print(tasks_flags)
        print(tasks_times)
        measure_cores_load(cores_loads, T_list, Launch_dependency, Cores_tasks_list, tasks_times, tasks_ticks, count_10ms_slots, time_start_ref_40, 4)
        print("At this point, Cores loading is", cores_loads)
        [input_binding_ctr, buffer_release_time, buffer_consume_time, in_buf_cons_lat] = measure_in_buffer_valid_lat(input_buffer_release_task, input_buffer_consume_task, buffer_release_time, input_binding_ctr, in_buf_cons_lat, tasks_times)
        print("At, this point, Input buffers duration stats are", in_buf_cons_lat)
        launch_list = get_task(0, T_list, 19)
        print(launch_list)
        task_idx = 0
        [task, core] = task_to_launch(launch_list, task_idx)
        cores_active[core] = 1
        print(cores_active)
        waiting_tasks = determine_waiting_tasks(core, launch_list, tasks_flags, tasks_times, tasks_ticks, t_10ms_flag, t_40ms_flag, t[core])
        print(waiting_tasks)
        if (waiting_tasks[0] == 0):
            mod_flag = 1
            time_adjust = 0
            for j in range(4):
                if count_10ms_slots[j] % 4 != 0:
                    mod_flag = 0
            if mod_flag == 1:
                time_adjust = max(tasks_times) - count_10ms_slots[0] * 10000
                if time_adjust < 0:
                    time_adjust = 0
            print("Need to adjust time by", time_adjust)
            t[core] = t[core] + time_adjust
            next_time_events = determine_next_time_events(core, t[core], launch_list, tasks_flags, tasks_times, tasks_ticks, t_10ms_flag, t_40ms_flag)
            tasks_times = set_all_tasks_times_to_next_tick(tasks_times, tasks_ticks, 19, 10)
            if mod_flag == 1 and time_adjust == 0:
                tasks_times = set_all_tasks_times_to_next_tick(tasks_times, tasks_ticks, 19, 40)
                t[1] = tasks_times[Cores_tasks_list[1][-1]]
            print(tasks_times)
            print(next_time_events)
            print("At time =", t[core])
            tasks_flags = launch_tasks(launch_list, core, tasks_flags, tasks_times, tasks_ticks, t[0], t[core], 1)
            time_backup = t[0]
            [t1, c, idx, task] = go_next(next_time_events)
            print("New time =", t1, ", core to free up =", c, ", task to terminate =", task, ", Event to cancel =", idx)
            time_map[core, int(t[core] / 100)] = 10 * task + Tasks_offset_color[task]
            time_map[core, int(t1 / 100) - 1] = 10 * task + Tasks_offset_color[task]
            #print("Gfx data:", core, int(t[core] / 100), time_map[core, int(t[core] / 100)])
            index = int(t[core] / 100) + 1
            while index < (int(t1 / 100) - 1):
                time_map[core, index] = 10 * task + Tasks_offset_color[task]
                #print("Gfx data:", core, index, time_map[core, index])
                index = index + 1
            #if int(t1 / 100) - 1 > int(t[core] / 100):
            #    print("Gfx data:", core, int(t1 / 100) - 1, time_map[core, int(t1 / 100) - 1])
            delta_t = t1 - t[0] + time_adjust
            t[c] = t1
            if count_10ms_slots[0] % 4 == 0:
                time_start_ref_40 = t1
            print(cores_active)
            for i in range(4):
                if cores_active[i] == 0 and (t_10ms_flag[i] == 1 or (count_10ms_slots[0] % 4 == 0 and t_40ms_flag[i] == 1)):
                    if i != 1 or time_adjust == 0:
                        t[i] = t[i] + delta_t
                    if i == 1 and (count_10ms_slots[0] % 4 == 0 and t_40ms_flag[i] == 1 and time_adjust != 0):
                        t[i] = t[0]
            print(t)
            next_time_events = clean_events(next_time_events, idx)
            print(next_time_events)
            cores_active[c] = 0
            tasks_flags = close_task(task, tasks_flags)
            tasks_times[task] = t1
            print(tasks_flags)
            print(tasks_times)
            print(task)
            if goto[0] == 1:
                task0 = task
                goto[0] = 0
                tasks_flags = clear_all_core_flags(Cores_tasks_list[0], tasks_flags)
            if goto[1] == 1:
                task1 = task
                goto[1] = 0
                tasks_flags = clear_all_core_flags(Cores_tasks_list[1], tasks_flags)
            if goto[2] == 1:
                task2 = task
                goto[2] = 0
                tasks_flags = clear_all_core_flags(Cores_tasks_list[2], tasks_flags)
            if goto[3] == 1:
                task3 = task
                goto[3] = 0
                tasks_flags = clear_all_core_flags(Cores_tasks_list[3], tasks_flags)
            tasks_flags[1] = 1
            print(tasks_flags)
        else:
            print("Infinite Loop!")
            cores_active[core] = 0
            break
```
